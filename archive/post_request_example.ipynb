```python
import folium
import requests as r
import json

def geocoder_conn_str(input_addr):

    QUERY = '%20'.join(input_addr.split(' '))
    BASE = 'https://us1.locationiq.com/v1/search.php?key='

    conn_str = BASE + 'eeb5b8923de08c' + '&q=' + QUERY + '&format=json'

    data = r.get(conn_str).json()

    return [float(data[0]['lat']), float(data[0]['lon'])]
```

```python
except_list = []

for index, row in df.iterrows():
    try:
        data = geocoder_conn_str(row['CTYST'])
        df.set_value(index, 'lat', data[0])
        df.set_value(index, 'lon', data[1])
        print('Completed Index: {} & City: {}'.format(index, row['CTYST']))
    except:
        except_list.append(index)
        print('Error: {} & City: {}'.format(index, row['CTYST']))
except_list
```
